```python
def auto_size(df, plot_cols):
    import matplotlib.pyplot as plt
    import matplotlib.patches as mpatches
    for col in plot_cols:
        fig = plt.figure(figsize=(10, 10))
        ax = fig.gca()
        temp1 = df.ix[df['Food you like most'] == 'b']       
        temp2 = df.ix[df['Food you like most'] == 'w']
      
        if temp1.shape[0] > 0:                    
            temp1.plot(kind = 'scatter', x = col, y = 'Favorite transport mode' , 
                           ax = ax, color = 'DarkRed', s= 500 * FinalDataset4['Place of interest'], alpha = 0.7)                          
        if temp2.shape[0] > 0:                    
            temp2.plot(kind = 'scatter', x = col, y = 'Favorite transport mode' , 
                           ax = ax, color = 'DarkBlue', s= 500 * FinalDataset4['Place of interest'], alpha = 0.7) 
    ax.set_title('Scatter plot of Favourite food VS. ' + col + ' Wtih Range of Favourite place.')
    red_patch = mpatches.Patch(color='DarkRed', label='bangladeshi')   
    blue_patch = mpatches.Patch(color='DarkBlue', label='western')
    plt.legend(handles=[red_patch, blue_patch])
    return plot_cols
auto_size(FinalDataset4, ['Sleep duration']) 
```
